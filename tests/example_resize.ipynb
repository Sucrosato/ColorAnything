```python
import cv2
import os
path = 'E:/work/Code/ColorAnything/examples'
```

```python
filelist = os.listdir(path)
filelist
```

```python
for file in filelist:
    img = cv2.imread(path+'/'+file)
    img = cv2.resize(img, (299, 299), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite(path+'_resized/' + file, img)
    print(path+'_resized/' + file)
```
