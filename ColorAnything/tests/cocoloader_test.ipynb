```python
import torch
import cv2
import numpy as np
def cocoload(img_path):
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (518, 518), interpolation=cv2.INTER_CUBIC)
    img = img.astype(np.float32) / 255.0
    img = (img - mean) / std

    grey = np.sum(img, axis=-1) / 3.0
    grey = torch.from_numpy(grey).float()
    label = torch.from_numpy(img).permute(2, 0, 1).float()

    return grey, label
```

```python
img_path = r'E:\work/Code/ColorAnything/examples/000000001168.jpg'
x, y = cocoload(img_path)
```
